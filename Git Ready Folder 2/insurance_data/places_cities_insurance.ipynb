```python
# Read in the data file with the city/town/city-county names

ga_cities = pd.read_csv("../census_data/ga_cities.csv")
```

```python
ga_cities_geo = ga_cities.copy()
```

```python
ga_cities_geo["latitude"] = ""
ga_cities_geo["longitude"] = ""
ga_cities_geo["county"] = ""
```

```python
params = {"key": gkey}

# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"
```

```python
for index, row in ga_cities_geo.iterrows():

    # get name of city
    city = row['name']

    # add keyword to params dict
    params['address'] = f"{city}, Georgia"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {city}, Georgia.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Trying to get a coordinate for {city}, Georgia.")
        ga_cities_geo.loc[index, 'latitude'] = results[0]["geometry"]["location"]["lat"]
        ga_cities_geo.loc[index, 'longitude'] = results[0]["geometry"]["location"]["lng"]
        ga_cities_geo.loc[index, 'county'] = results[0]["address_components"][1]["long_name"]
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

**OK now for the full set of cities!**

```python
agencies_df = pd.DataFrame({"agency_name":[],
                            "agency_status":[],
                            "agency_address":[],
                            "compound_code":[],
                            "agency_latitude":[],
                            "agency_longitude":[],
                            "agency_target_city":[], 
                            "agency_target_county":[]})

display(agencies_df)
```

```python
# find the closest insurance agencies to each city coordinate

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {
    "radius": 50000,
    "type": "insurance_agency",
    "key": gkey,
}

local_count = 0

# use iterrows to iterate through pandas dataframe
for index, row in ga_cities_geo.iterrows():

    lat = row['latitude']
    lon = row['longitude']
    city_name = row['name']
    county_name = row['county']

    # add location to params dict
    params['location'] = f"{lat}, {lon}"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {city_name}.")
    response = requests.get(base_url, params = params).json()
    
    # extract results
    results = response['results']

    for i in range(len(response["results"]) - 1):
        
        try:
            agencies_df.loc[local_count, "agency_name"] = results[i]["name"]
            agencies_df.loc[local_count, "agency_status"] = results[i]["business_status"]
            agencies_df.loc[local_count, "agency_address"] = results[i]["vicinity"]
            agencies_df.loc[local_count, "compound_code"] = response["results"][i]["plus_code"]["compound_code"]
            agencies_df.loc[local_count, "agency_latitude"] = results[i]["geometry"]["location"]["lat"]
            agencies_df.loc[local_count, "agency_longitude"] = results[i]["geometry"]["location"]["lng"]
            agencies_df.loc[local_count, "agency_target_city"] = city_name
            agencies_df.loc[local_count, "agency_target_county"] = county_name
        
        except (KeyError, IndexError):
            print("Missing field/result... skipping.")
        
        local_count += 1
    print("------------")
```

```python
display(agencies_df)
```

```python
# Pull out the target_city and target_county so that that duplicates can be dropped!

agencies_df_no_targets = agencies_df.drop(["agency_target_city", "agency_target_county"], axis = 1)

# check
agencies_df_no_targets.head()
```

```python
agencies_df_culled = agencies_df_no_targets.drop_duplicates()
print(f"The number of unique agencies, after dropping duplicates is {len(agencies_df_culled)}")
```
